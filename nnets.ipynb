### Neural Nets

We also generated a neural net for our startup data using the neural net toolbox from Matlab. We imported the scaled features data into Matlab and log(funding) data. After this, we trained a neural net using a 15% of the data as a validation, a 15% of the data as testing data, and 70% of the data as train data. The neural net was trained using their implementation of Bayesian Regularization. MSE was used as the performance metric. A middle layer with approximately half the total number of nodes (50 nodes) was chosen. An image of the neural net architecture is below.
