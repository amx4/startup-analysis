```python
def trydetect(tweet):
    try:
        a = translator.detect_language(tweet)
    except:
        return False
    else:
        return a

lang = translator.get_languages()
def translateSeries(srs):
    tweet = str(srs['Text'])
    if trydetect(tweet) in lang:
        trans_text = trans(translator, tweet)
        while trans_text == False:
            global currnum
            currnum += 1
            if currnum >= len(keys):
                currnum = 0
            cid, kk = keys[currnum]
            global translator
            translator = Translator(cid, kk)
            trans_text = trans(translator, tweet)
        srs['Text'] = trans_text
    return srs
```
